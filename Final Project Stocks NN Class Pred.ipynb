```python
# Pandas and numpy for data manipulation
import pandas as pd
import numpy as np

# Imputing missing values and scaling values
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

#Neural Network
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Read in data into dataframes 
train_features = pd.read_csv('data/training_features2.csv')
test_features = pd.read_csv('data/testing_features2.csv')
train_labels = pd.read_csv('data/training_labels2.csv')
test_labels = pd.read_csv('data/testing_labels2.csv')

# Display sizes of data
print('Training Feature Size: ', train_features.shape)
print('Testing Feature Size:  ', test_features.shape)
print('Training Labels Size:  ', train_labels.shape)
print('Testing Labels Size:   ', test_labels.shape)
```

```python
train_features.head()
```

```python
# Create an imputer object with a median filling strategy
imputer = SimpleImputer(strategy='median')

# Train on the training features
imputer.fit(train_features)

# Transform both training data and testing data
X = imputer.transform(train_features)
X_test = imputer.transform(test_features)
```

```python
print('Missing values in training features: ', np.sum(np.isnan(X)))
print('Missing values in testing features:  ', np.sum(np.isnan(X_test)))
```

```python
# Make sure all values are finite
print(np.where(~np.isfinite(X)))
print(np.where(~np.isfinite(X_test)))
```

```python
X_scaler = MinMaxScaler().fit(X)
X_train_scaled = X_scaler.transform(X)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(train_labels)
encoded_y_train = label_encoder.transform(train_labels)
encoded_y_test = label_encoder.transform(test_labels)
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

# Neural Network

```python
model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=654))
model.add(Dense(units=6, activation='relu'))
model.add(Dense(units=6, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=200,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaled[:5])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:5])}")
```
