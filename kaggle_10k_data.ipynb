```python
allFilePaths = []

for i in range(4, 9):
    #get year in format used in file name
    year = i
      
    #create filename dynamically
    subFolder = "Data_2/"
    fileName = f"{subFolder}201{year}_Financial_Data.csv"
    allFilePaths.append(fileName)
    
#Print all file paths
allFilePaths
```

```python
df = pd.DataFrame()
year_column = 2014
for path in allFilePaths:
    loop_df = pd.read_csv(path)
    loop_df = loop_df.rename(columns={'Unnamed: 0': 'Ticker'})
    loop_df['Year'] = year_column
    df = df.append(loop_df)
    year_column += 1
df.info()
```
